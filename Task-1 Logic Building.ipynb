```python
# Task-2
# Student marks
marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

# Analyze results
for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

# Final result
print("Pass Students:", pass_count)
print("Fail Students:", fail_count)
```

```python
# Task-3
# Original names list
names = [" Alice ", "bob", " CHARLIE "]

# Clean data
cleaned_names = []
for name in names:
    cleaned_name = name.strip().lower()
    cleaned_names.append(cleaned_name)

# Print cleaned list
print("Cleaned Names:", cleaned_names)
```

```python
# Task-4
# Messages list
messages = ["Hi", "Welcome to the platform", "OK"]

# Analyze message lengths
for msg in messages:
    length = len(msg)
    print(f"Message: '{msg}' | Length: {length}")
    
    if length > 10:
        print("→ This message is longer than 10 characters")
```

```python
# Task-5
# System logs
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

# Count ERROR entries
error_count = 0
for log in logs:
    if log == "ERROR":
        error_count += 1

# Print result
print("Total ERROR messages:", error_count)
```
