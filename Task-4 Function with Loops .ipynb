```python
#Problem-01
def validate_recharge():
    valid_plans = [199, 299, 399, 599]

    while True:
        amount = int(input("Enter recharge amount: "))

        if amount < 50:
            print("Amount must be at least ₹50. Try again.")
            continue

        if amount not in valid_plans:
            print("Invalid plan. Valid plans are 199, 299, 399, 599. Try again.")
            continue

        print("Recharge successful!")
        break

validate_recharge()
```

```python
#Problem-2
def check_inventory(stock):
    for product, quantity in stock.items():
        if quantity < 15:
            print(f"{product}: Reorder Alert")
        else:
            print(f"{product}: Stock OK")

inventory = {
    "Apples": 10,
    "Bananas": 20,
    "Oranges": 5,
    "Mangoes": 18
}

check_inventory(inventory)
```

```python
#Problem-3
def process_result(marks):
    total = 0

    for m in marks:
        total += m

    average = total / len(marks)

    if average >= 50:
        print("Pass")
    else:
        print("Fail")

marks_list = [60, 55, 40, 70]

process_result(marks_list)
```

```python
#Problem-4
def calculate_fare(distance, peak):
    fare = 50 + (12 * distance)
    if peak.lower() == "yes":
        fare *= 1.25
    return fare

while True:
    distance = float(input("Enter distance (km): "))
    peak = input("Is it peak hour? (yes/no): ")

    total_fare = calculate_fare(distance, peak)
    print("Total Fare: ₹", round(total_fare, 2))

    retry = input("Do you want to calculate again? (yes/no): ")
    if retry.lower() != "yes":
        break
```

```python
#Problem-5
def check_eligibility(attendance):
    present = 0

    for status in attendance:
        if status == "P":
            present += 1

    percentage = (present / len(attendance)) * 100

    if percentage >= 75:
        return "Eligible"
    else:
        return "Not Eligible"

attendance_list = ["P", "P", "A", "P", "P", "A", "P"]

result = check_eligibility(attendance_list)
print(result)
```

```python
#Problem-6
def check_password_strength(password):
    has_digit = False
    has_special = False

    for ch in password:
        if ch.isdigit():
            has_digit = True
        if ch in "@#$":
            has_special = True

    if len(password) >= 8 and has_digit and has_special:
        return "Strong Password"
    else:
        return "Weak Password"

pwd = input("Enter password: ")
print(check_password_strength(pwd))
```
