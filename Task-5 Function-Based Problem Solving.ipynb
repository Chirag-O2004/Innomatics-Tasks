```python
#Problem-1
def analyze_engagement(likes_list):
    total_likes = 0

    for likes in likes_list:
        total_likes += likes

    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    print("Total Likes:", total_likes)
    print("Post Status:", status)

likes_data = [200, 300, 150, 400]

analyze_engagement(likes_data)
```

```python
#Problem-2
def check_medicine_stock(stock):
    if stock < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"

    print("Medicine Stock:", stock)
    print("Status:", status)

check_medicine_stock(6)
```

```python
#Problem-3
def rainfall_checker(rainfall_data, required_level):
    total = 0

    for r in rainfall_data:
        total += r

    average = total / len(rainfall_data)

    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print("Average Rainfall:", average)
    print("Rainfall Status:", status)

rain_data = [70, 75, 68, 80, 67]
required = 70

rainfall_checker(rain_data, required)
```

```python
#Problem-4
def detect_duplicates(usernames):
    if len(usernames) != len(set(usernames)):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")

users = ["alice", "bob", "charlie", "alice"]

detect_duplicates(users)
```

```python
#Problem-5
def check_appointment_eligibility(age):
    if age >= 18:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Patient Age:", age)
    print("Eligibility Status:", status)

check_appointment_eligibility(21)
```

```python
#Problem-6
def filter_premium_crops(prices, threshold=2000):
    premium = []

    for price in prices:
        if price > threshold:
            premium.append(price)

    print("Premium Crops:", premium)

crop_prices = [1500, 2500, 1800, 3200, 900]

filter_premium_crops(crop_prices)
```

```python
#Problem-7
def check_app_health(errors):
    if errors == 0:
        status = "Healthy"
    elif errors <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print("Error Count:", errors)
    print("System Status:", status)

check_app_health(7)
```

```python
#Problem-8
def check_transaction(amount):
    limit = 50000

    if amount <= limit:
        status = "Approved"
    else:
        status = "Rejected"

    print("Transaction Amount:", amount)
    print("Transaction Status:", status)

check_transaction(60000)
```

```python
#Problem-9
def check_exam_eligibility(attendance):
    present = 0

    for status in attendance:
        if status == "P":
            present += 1

    percentage = (present / len(attendance)) * 100

    if percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"

    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", eligibility)

attendance_list = ["P", "P", "A", "P", "P"]

check_exam_eligibility(attendance_list)
```

```python
#Problem-10
def electricity_bill(units):
    bill = 0

    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + (units - 100) * 5
    else:
        bill = (100 * 3) + (100 * 5) + (units - 200) * 7

    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    return bill, usage

units_consumed = 250

bill_amount, usage_status = electricity_bill(units_consumed)

print("Total Bill:", bill_amount)
print("Usage Level:", usage_status)
```
