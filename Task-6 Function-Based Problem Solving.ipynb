```python
#Problem-1
def manage_parking(capacity, logs):
    current = 0
    peak = 0

    for log in logs:
        if log == "IN":
            current += 1
        elif log == "OUT" and current > 0:
            current -= 1

        if current > peak:
            peak = current

    if current > capacity:
        status = "Over Capacity Alert"
    else:
        status = "Available"

    print("Currently Parked Vehicles:", current)
    print("Parking Status:", status)
    print("Peak Parking Usage:", peak)

capacity = 50
vehicle_logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

manage_parking(capacity, vehicle_logs)
```

```python
#Problem-2
def estimate_delivery_time(distance, traffic, weather):
    base_time = distance * 5

    if traffic.lower() == "high":
        base_time += 10
    elif traffic.lower() == "medium":
        base_time += 5

    if weather.lower() == "rainy":
        base_time += 5
    elif weather.lower() == "stormy":
        base_time += 15

    print("Estimated Delivery Time:", int(base_time), "minutes")

estimate_delivery_time(8, "High", "Rainy")
```

```python
#Problem-3
def analyze_occupancy(total_seats, booked_seats):
    booked_count = len(booked_seats)
    occupancy = (booked_count / total_seats) * 100

    if occupancy == 100:
        status = "Housefull"
    elif occupancy >= 75:
        status = "Almost Full"
    else:
        status = "Available"

    print("Occupancy:", f"{occupancy:.0f}%")
    print("Show Status:", status)

total = 200
booked = [1] * 150

analyze_occupancy(total, booked)
```

```python
#Problem-4
def classify_server_load(cpu_readings):
    total = 0

    for reading in cpu_readings:
        total += reading

    average = total / len(cpu_readings)

    if average < 50:
        status = "Normal"
    elif average <= 80:
        status = "Warning"
    else:
        status = "Critical"

    print("Average CPU Load:", f"{average:.0f}%")
    print("Server Status:", status)

readings = [45, 60, 70, 85, 90]

classify_server_load(readings)
```

```python
#Problem-5
def monitor_resources(usage):
    overused = []

    for resource, hours in usage.items():
        if hours > 8:
            overused.append(resource)

    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")

resource_usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}

monitor_resources(resource_usage)
```

```python
#Problem-6
def manage_registrations(capacity, registrations):
    confirmed = min(registrations, capacity)
    waitlisted = max(0, registrations - capacity)

    if registrations >= capacity:
        status = "Closed"
    else:
        status = "Open"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)

manage_registrations(100, 105)
```
