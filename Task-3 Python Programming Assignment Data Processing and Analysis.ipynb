```python
#Problem-01
# Dictionary containing employee names and their performance scores
employees = {
    "Ravi": 92,
    "Anita": 88,
    "Kiran": 92,
    "Suresh": 85
}

# Find the highest performance score
max_score = max(employees.values())

# Find all employees who have the highest score (handle ties)
top_performers = [name for name, score in employees.items() if score == max_score]

# Display the result
print("Top Performers Eligible for Bonus:", ", ".join(top_performers), f"(Score: {max_score})")
```

```python
#Problem-02
# Search query input
query = "Buy mobile phone buy phone online"

# Convert to lowercase
query = query.lower()

# Remove punctuation
import string
query = query.translate(str.maketrans("", "", string.punctuation))

# Split into words
words = query.split()

# Count frequency
freq = {}
for word in words:
    if word in freq:
        freq[word] += 1
    else:
        freq[word] = 1

# Filter keywords with frequency > 1
result = {word: count for word, count in freq.items() if count > 1}

# Display result
print(result)
```

```python
#Problem-03
# Sensor readings list (index represents hour)
sensor_readings = [3, 4, 7, 8, 10, 12, 5]

# List to store valid readings as (hour, value)
valid_readings = []

# Check each reading
for hour, value in enumerate(sensor_readings):
    # Keep only even readings
    if value % 2 == 0:
        valid_readings.append((hour, value))

# Display result
print("Valid Sensor Readings (Hour, Value):")
print(valid_readings)
```

```python
#Problem-04
# List of employee emails
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

# Dictionary to count domain usage
domain_count = {}

# Extract domain and count
for email in emails:
    domain = email.split("@")[1]
    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

# Calculate and print percentage usage
total = len(emails)

for domain, count in domain_count.items():
    percentage = (count/total) * 100
    print(f"{domain}: {percentage:.0f}%")
```

```python
#Problem-05
# Daily sales data
sales = [1200, 1500, 900, 2200, 1400, 3000]

# Calculate average sales
average = sum(sales) / len(sales)

# Threshold for spike (30% above average)
threshold = average * 1.3

# Detect and display spikes
for day, value in enumerate(sales, start=1):
    if value > threshold:
        print(f"Day {day}: {value}")
```

```python
#Problem-06
# List of user IDs
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Dictionary to count occurrences
count = {}

for uid in user_ids:
    if uid in count:
        count[uid] += 1
    else:
        count[uid] = 1

# Display duplicates only
for uid, c in count.items():
    if c > 1:
        print(f"{uid} → {c} times")
```
